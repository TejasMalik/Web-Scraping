```python
import bs4
import requests
```

```python
url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
response = requests.get(url)
```

```python
response.content
```

```python
soup = bs4.BeautifulSoup(response.content)
```

```python
article_element = soup.find_all('img')
```

```python
article_element[7]
```

```python
article = article_element[7]
```

```python
article.attrs['src']
```

```python
with open('insipiration.jpg', 'wb') as file:
    img_url = article.attrs['src']
    response = requests.get(img_url)
    
    file.write(response.content)
```
